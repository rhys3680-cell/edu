## 짝수와 홀수 - 나머지 연산

- 짝수는 "2로 나누어떨어지는 수".
- 홀수는 "2로 나누었을 때, 나머지가 1이 되는 수

```python
print(4 % 2 == 0)
print(5 % 2 == 0)
```

```python
def is_even(n):
    return n % 2 == 0

print(is_even(4))
print(is_even(5))
```

```python
def divisors(n):
    return [i for i in range(1, n + 1) if n % i == 0]

divisors(12)
```

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, n):
        if n % i == 0:
            return False

    return True

is_prime(11)
```

```python
def gcd(a, b):
    if b == 0: 
        return a
    return gcd(b, a % b)

print(gcd(13, 11))
print(gcd(12, 6))
print(gcd(12, 18))
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(gcd(13, 11))
print(gcd(12, 6))
print(gcd(12, 18))
```

```python
import pandas as pd

# 유클리드 호제법이 (a, b) → (b, a%b)로 줄어드는 과정
def gcd_steps(a, b):
    rows = []
    step = 1
    while b:
        rows.append({"단계": step, "a": a, "b": b, "a % b": a % b})
        a, b = b, a % b
        step += 1
    rows.append({"단계": step, "a": a, "b": b, "a % b": "-"})  # b=0 → 종료
    return pd.DataFrame(rows)

gcd_steps(48, 18)   # b가 0이 되는 순간의 a(=6)가 gcd
```

```python
def lcm(a, b):
    return a * b // gcd(a, b)

print(lcm(4, 6))    # 12
print(lcm(12, 18))  # 36
```
