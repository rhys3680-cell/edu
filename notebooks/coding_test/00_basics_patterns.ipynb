# 유형 00 — Lv.0 문법 패턴

> **대상**: 문법(`for`/`if`/리스트/문자열)은 아는데, "문제를 보면 어떻게 코드로 옮길지"가 막히는 단계.
> **목표**: 자료구조·알고리즘 없이, **문제 지문 → 코드 구조 번역**을 손에 익히기.

핵심은 새 문법을 배우는 게 아니라, **"이런 말이 나오면 이런 코드"**라는 패턴 매칭을 몸에 새기는 것입니다.

## 패턴 1 — "모든 ~에 대해 무언가 한다" → 반복문

지문에 **"각각", "모든", "1부터 n까지"**가 나오면 `for`입니다.

```python
# "1부터 n까지의 합을 구하라"
def solution(n):
    total = 0
    for i in range(1, n + 1):   # 1, 2, ..., n
        total += i
    return total
```

> 번역 규칙: **"~까지" → `range`**, **"누적/합/개수" → 루프 밖에 변수 만들고 루프 안에서 갱신**.

````python
def solution(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
````

````python
print(solution(10))
````

## 패턴 2 — "조건을 만족하는 것만" → 반복 + if

**"~인 경우만", "짝수만", "3보다 큰"** → 루프 안에 `if`.

```python
# "리스트에서 짝수만 더하라"
def solution(numbers):
    total = 0
    for x in numbers:
        if x % 2 == 0:          # 짝수 조건
            total += x
    return total
```

같은 걸 한 줄로 (익숙해지면):

```python
def solution(numbers):
    return sum(x for x in numbers if x % 2 == 0)
```

> 가르칠 포인트: 먼저 **풀어쓴 for+if**로 이해 → 그 다음 **컴프리헨션**으로 압축. 순서가 중요합니다.


````python
def solution(numbers):
    total = 0
    for x in numbers:
        if x % 2 == 0:
            total += x
    return total
````

````python
li1 = []

for i in range(101):
    li1.append(i)

print(li1)
print(solution(li1))
````

````python
def solution(numbers):
    return sum(x for x in numbers if x % 2 == 0)
````

````python
print(li1)
print(solution(li1))
````

## 예제 풀이

각 문제를 **① 지문에서 신호 찾기 → ② 치트시트로 패턴 분류 → ③ 코드 작성** 순서로 풉니다.

### 예제 1 — 두 정수 사이의 합

> 두 정수 `a`, `b`가 주어질 때, `a`와 `b` 사이에 속한 모든 정수의 합을 구하라. (`a ≤ b`가 아닐 수도 있음)

**① 신호**: "사이에 속한 모든 정수", "합"
**② 패턴**: 패턴 1 (모든 ~에 대해 → 반복) + 패턴 3 (합 → 누적 변수)
**③ 작성**:

```python
def solution(a, b):
    # a가 b보다 클 수도 있으니 작은 값/큰 값을 정리
    if a > b:
        a, b = b, a
    total = 0
    for i in range(a, b + 1):   # a부터 b까지 (b 포함이라 +1)
        total += i
    return total
```

> 함정 포인트: **`a > b`인 경우**를 빠뜨리면 `range`가 빈 범위가 되어 0이 나옴. "순서가 보장 안 된다"는 한 줄을 놓치지 않는 훈련.
>
> 공식을 알면 한 줄: `return (a + b) * (abs(b - a) + 1) // 2` — 하지만 처음엔 루프로 원리부터.

````python
def solution(a, b):
    if a > b:
        a, b = b, a
    total = 0
    for i in range(a, b + 1):
        total += i
    return total
````

````python
print(solution(1, 10))
````

````python
print(solution(10, 1))
````

````python
def solution(a, b):
    return (a + b) * (abs(b - a) + 1) // 2
````

````python
solution(10, 1)
````

> 정수 `n`이 주어질 때, `n` 이하의 짝수를 모두 더한 값을 반환하라.

**① 신호**: "n 이하의", "짝수를", "모두 더한"
**② 패턴**: 패턴 1 (반복) + 패턴 2 (조건만) + 패턴 3 (합)
**③ 작성**:

```python
def solution(n):
```

````python
def solution(n):
    total = 0
    for i in range(1, n + 1):
        if i % 2 == 0:
            total += i
    return total
````

### 예제 3 — 문자열 뒤집기

> 문자열 `s`가 주어질 때, 순서를 뒤집은 문자열을 반환하라.

**① 신호**: "문자열을", "뒤집은"
**② 패턴**: 패턴 4 (문자열 다루기) + 패턴 5 (각 글자를 순회)
**③ 작성** — 원리부터:

```python
def solution(s):
    result = ""
    for ch in s:
        result = ch + result    # 새 글자를 '앞에' 붙이면 역순이 됨
    return result
```

> 슬라이싱을 알면 한 줄: `return s[::-1]` (`[::-1]`은 "끝에서부터 거꾸로"). 원리(앞에 붙이기 누적)를 먼저 이해한 뒤 슬라이싱으로 압축하는 순서가 좋습니다.

````python
def solution(s):
    result = ""
    for ch in s:
        result = ch + result
    return result
````

````python
s = "hello world"

print(solution(s))
````

````python
def solution(s):
    return s[::-1]
````

````python
print(solution(s))
````

### 예제 4 — 자릿수 더하기

> 정수 `n`이 주어질 때, 각 자리 숫자의 합을 구하라. (예: `123` → `1 + 2 + 3 = 6`)

**① 신호**: "각 자리 숫자의", "합"
**② 패턴**: 패턴 4 (문자열로 변환) + 패턴 1·3 (순회하며 합)
**③ 작성** — 문자열로 바꿔 한 글자씩:

```python
def solution(n):
    total = 0
    for ch in str(n):           # 숫자를 문자열로 → 한 글자씩 순회
        total += int(ch)        # 글자를 다시 정수로 바꿔 더함
    return total
```

> 핵심 아이디어: **"각 자리"가 막막하면 `str(n)`으로 바꿔서 글자 순회로 바꾼다.** 숫자 ↔ 문자열 변환은 Lv.0의 단골 무기.
>
> 수학으로 푸는 법(나머지 연산)도 있지만, 문자열 변환이 더 직관적입니다:
> ```python
> def solution(n):
>     total = 0
>     while n > 0:
>         total += n % 10        # 마지막 자리
>         n //= 10               # 마지막 자리 버리기
>     return total
> ```

````python
def solution(n):
    total = 0
    for ch in str(n):
        total += int(ch)
    return total
````

````python
print(solution(1048587230))
````

````python
def solution(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
````

````python
print(solution(204923))
````
