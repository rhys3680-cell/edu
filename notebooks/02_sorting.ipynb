# 유형 02 — 정렬 (Sorting)

> **대상**: 자료구조 기초(스택/큐/해시)를 익힌 단계.
> **목표**: 정렬 알고리즘을 **구현**하는 게 아니라, **"무엇을 기준으로(`key`) 정렬하느냐"**를 설계하는 능력 기르기.

코딩테스트에서 정렬은 버블/퀵 정렬을 짜는 문제가 아닙니다. Python의 `sorted`/`sort`는 이미 빠르고(O(n log n)) 안정 정렬입니다. **승부는 `key`를 어떻게 잡느냐**에서 납니다.


## 1) 기본 — sorted vs sort

```python
nums = [3, 1, 2]

sorted(nums)          # [1, 2, 3]  ← 새 리스트 반환 (원본 유지)
nums.sort()           # 원본을 직접 정렬 (반환값 None)

sorted(nums, reverse=True)   # [3, 2, 1]  내림차순
```

> 규칙: **원본을 남겨야 하면 `sorted()`**, 그 자리에서 바꿔도 되면 `list.sort()`. 코테에선 보통 `sorted()`가 안전합니다.

````python
nums = [3, 1, 2]

print(sorted(nums))
nums.sort()

print(nums)

sorted(nums, reverse=True)
````

## 2) 핵심 — key로 정렬 기준 바꾸기

`key`는 **"각 원소를 무엇으로 보고 비교할지"** 정하는 함수입니다.

```python
words = ["banana", "kiwi", "apple"]

sorted(words, key=len)               # ['kiwi', 'apple', 'banana']  (길이순)
sorted(words, key=lambda w: w[-1])   # 마지막 글자순
```

**튜플 key — 다중 정렬 기준 (1순위, 2순위)**

```python
people = [("kim", 30), ("lee", 25), ("park", 30)]

# 나이 오름차순, 같으면 이름 오름차순
sorted(people, key=lambda x: (x[1], x[0]))
# [('lee', 25), ('kim', 30), ('park', 30)]
```

> 핵심: **key가 튜플이면 "앞 원소 먼저, 같으면 뒤 원소로"** 비교합니다. "A 기준으로 정렬하되 같으면 B 기준" 류 문제의 정답 패턴.

**일부만 내림차순으로 섞기**

```python
# 점수는 내림차순, 같으면 이름 오름차순 (한쪽만 역순)
sorted(people, key=lambda x: (-x[1], x[0]))   # 숫자는 부호 반전으로 역순
```

> 숫자 기준 하나만 내림차순으로 하고 싶을 때 `-값` 트릭을 자주 씁니다 (`reverse=True`는 전체에 걸리므로).


````python
words = ["banana", "kiwi", "apple"]

print(sorted(words, key=len))
print(sorted(words, key=lambda w: w[-1]))
````

````python
people = [("kim", 30), ("lee", 25), ("park", 30)]

print(sorted(people, key=lambda x: (x[1], x[0])))
print(sorted(people, key=lambda x: (-x[1], x[0])))
````

## 대표 문제 풀이 — 가장 큰 수 (프로그래머스 42746)

> 정수 배열 `numbers`를 이어 붙여 만들 수 있는 가장 큰 수를 반환하라. (예: `[6, 10, 2]` → `"6210"`)

**① 신호**: "이어 붙여서 가장 큰", "순서를 정해서 최대/최소"
**② 패턴**: 정렬 — 단, **비교 기준이 크기가 아니라 "이어 붙인 결과"**

**왜 단순 정렬이 안 되나**: `[3, 30]`을 크기순으로 두면 `"303"`인데, `"330"`이 더 큽니다. 즉 **두 수를 이어 붙여 비교**해야 합니다.

```python
def solution(numbers):
    # 문자열로 바꾸고, "두 개를 이어 붙인 결과"가 큰 순서로 정렬
    nums = list(map(str, numbers))
    nums.sort(key=lambda x: x * 3, reverse=True)   # x*3: 자릿수 맞춰 비교하는 트릭
    answer = "".join(nums)
    return answer if answer[0] != "0" else "0"     # [0,0,0] → "000" 방지
```

**검증:**

```python
print(solution([6, 10, 2]))       # "6210"
print(solution([3, 30, 34, 5, 9]))# "9534330"
print(solution([0, 0, 0]))        # "0"  (전부 0이면 "000"이 아니라 "0")

assert solution([6, 10, 2]) == "6210"
assert solution([3, 30, 34, 5, 9]) == "9534330"
assert solution([0, 0, 0]) == "0"
print("통과 ✅")
```

> 가르칠 포인트 두 가지:
> 1. **`key=lambda x: x*3`** — 숫자가 1000 이하라 3번 반복하면 자릿수가 충분히 맞춰져, 이어 붙였을 때 누가 앞에 와야 큰지를 문자열 비교로 판단할 수 있습니다. (`"3"*3="333"` vs `"30"*3="303030"` → `"333" > "303030"`이므로 3이 앞)
> 2. **`"0"` 엣지 케이스** — `[0,0,0]`이면 결과가 `"000"`이 되는데, 가장 앞이 `"0"`이면 통째로 `"0"`으로 바꿔줘야 합니다.


````python
def solution(numbers):
    nums = list(map(str, numbers))
    nums.sort(key=lambda x: x * 3, reverse=True)
    answer = "".join(nums)
    return answer if answer[0] != "0" else "0"
````

````python
print(solution([6, 10, 2]))
print(solution([3, 30, 34, 5, 9]))
print(solution([0, 0, 0]))

assert solution([6, 10, 2]) == "6210"
assert solution([3, 30, 34, 5, 9]) == "9534330"
assert solution([0, 0, 0]) == "0"
print("통과")
````
